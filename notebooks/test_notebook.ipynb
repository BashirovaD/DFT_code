```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
def mod_func(delta_omega: int, t: float):
  func = np.sin(delta_omega * t / 2) ** 2 / ( (delta_omega / 2) ** 2 * t)
  plt.plot(t, func)
  plt.xlabel('time')
  plt.show()
```

```python
mod_func(2, np.linspace(-10, 10, 5000))
```
